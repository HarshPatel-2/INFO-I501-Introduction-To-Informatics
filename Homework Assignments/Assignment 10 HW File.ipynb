```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

p_defective = 0.05
n = 20
x = np.arange(0, n+1)
probability = binom.pmf(x, n, p_defective)

plt.bar(x, probability, width=0.8, align='center')
plt.title("Probability of Defective Products in a Sample of 20")
plt.xlabel("Number of Defective Products")
plt.ylabel("Probability")
plt.show()
```

```python
from scipy.stats import poisson

average_packets_per_second = 15
time_interval = 2

lambda_parameter = average_packets_per_second * time_interval
probability_more_than_20 = 1 - poisson.cdf(20, lambda_parameter)

print("Probability of receiving more than 20 packets in a 2-second interval:", probability_more_than_20)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

np.random.seed(0)

sample_size = 300
rate_parameter = 0.2
random_sample = np.random.exponential(scale=1/rate_parameter, size=sample_size)

plt.hist(random_sample, bins=np.linspace(0, 15, 31), density=True, alpha=0.7, color='green')

x = np.linspace(0, 15, 1000)
pdf = expon.pdf(x, scale=1/rate_parameter)
plt.plot(x, pdf, 'r-', lw=2)

plt.title("Histogram and PDF of Exponential Distribution")
plt.xlabel("X (Random Variable) - Exponential Distribution")
plt.ylabel("Probability Density")

plt.show()
```

```python
import random

num_experiments = 300
num_successful_outcomes = [0] * 5 

total_cards = 52
odd_cards = [1, 3, 5, 7, 9]

for _ in range(num_experiments):
    deck = list(range(1, total_cards + 1))
    drawn_cards = random.sample(deck, 5)
    success = True
    for i in range(5):
        if drawn_cards[i] not in odd_cards:
            success = False
            break
    if success:
        num_successful_outcomes[4] += 1
    else:
        for i in range(5):
            if drawn_cards[i] in odd_cards:
                num_successful_outcomes[i] += 1

probabilities = [success / num_experiments for success in num_successful_outcomes]

for i, probability in enumerate(probabilities):
    print(f"Probability of getting an odd-numbered card on draw {i+1}: {probability:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 1000
std_dev = 100

x = np.linspace(800, 1200, 400) 
pdf = norm.pdf(x, loc=mean, scale=std_dev)

plt.plot(x, pdf, 'r-', lw=2)
plt.title("Probability Density Function (PDF) of Bulb Lifetimes")
plt.xlabel("Lifetime (hours)")
plt.ylabel("PDF")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

data = pd.read_csv('road-weather-information-stations.csv')

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

road_surface_temperatures = []
air_temperatures = []

num_samples = 10
bins = 20

for _ in range(100):
    sampled_data = data.sample(n=num_samples)
    road_surface_temperatures.extend(sampled_data['RoadSurfaceTemperature'])
    air_temperatures.extend(sampled_data['AirTemperature'])

axes[0].hist(road_surface_temperatures, bins=bins, color='magenta', alpha=0.7)
axes[0].set_title("Road Surface Temperature Histogram")
axes[0].set_xlabel("Temperature (°F)")
axes[0].set_ylabel("Frequency")

axes[1].hist(air_temperatures, bins=bins, color='cyan', alpha=0.7)
axes[1].set_title("Air Temperature Histogram")
axes[1].set_xlabel("Temperature (°F)")
axes[1].set_ylabel("Frequency")

fig.suptitle("Randomly Sampled Temperature Histograms")

plt.tight_layout()
plt.show()
```
