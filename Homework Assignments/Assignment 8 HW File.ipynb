```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
numbers = [0.2, 0.4, 1.5, 2.8, 3.0, 2.1, 1.9, 1.6, 1.8, 1.3, 1.4, 0.7]

plt.plot(months, numbers, label='GPA', linestyle='-', color='b')

plt.xlabel('Months')
plt.ylabel('Numerical Value (0-3.2)')
plt.title('Monthly GPA Data Trend')

plt.ylim(0, 3.2)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y_values = [0, 13, 8, 11, 9, 17, 9, 11, 25]
line_chart_data = [0, 14, 8, 11, 9, 17, 9, 11, 25]
scatter_plot_data = [2, 12, 5, 20, 14, 8, 16, 10, 25]
bar_chart_data = [0, 13, 8, 11, 9, 18, 9, 11, 25]

def create_histogram(ax):
    ax.hist(y_values, bins=7, color='blue', edgecolor='white')
    ax.set_title('Histogram')
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

def create_line_chart(ax):
    ax.plot(line_chart_data, color='red')
    ax.set_title('Line Chart')
    ax.set_xlabel('Index')
    ax.set_ylabel('Value')

def create_scatter_plot(ax):
    ax.scatter(range(len(scatter_plot_data)), scatter_plot_data, color='green')
    ax.set_title('Scatter Plot')
    ax.set_xlabel('Index')
    ax.set_ylabel('Value')

def create_bar_chart(ax):
    ax.bar(range(len(bar_chart_data)), bar_chart_data, color='orange')
    ax.set_title('Bar Chart')
    ax.set_xlabel('Index')
    ax.set_ylabel('Value')

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
chart_functions = [create_histogram, create_line_chart, create_scatter_plot, create_bar_chart]
for i, create_chart in enumerate(chart_functions):
    create_chart(axs[i // 2, i % 2])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Flavored milk whole', 'Eggnog and buttermilk', 'Plain 2% milk', 'Plain whole milk', 'Plain 1% milk', 'Skim milk']
percentages = [9, 1, 35, 27, 14, 14]
colors = ['pink', 'yellow', 'green', 'purple', 'brown', 'blue']

plt.figure(figsize=(6, 6))
plt.pie(percentages, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title("U.S. Per Capita Fluid Milk Availability, Gallons, Shares, 2013")

plt.axis('equal') 
plt.show()
```

```python
import numpy as np

matrix_X = np.array([[4, 3, 1], [2, 6, 7]])
matrix_Y = np.array([[4, 3], [8, 4], [7, 6]])
result = np.dot(matrix_X, matrix_Y)

print("Matrix X:")
print(matrix_X)
print("\nMatrix Y:")
print(matrix_Y)
print("\nResult of Matrix Multiplication (X * Y):")
print(result)
```

```python
import MySQLdb
import matplotlib.pyplot as plt

db = MySQLdb.connect(
    host="localhost",
    user="hrp1",
    password="inferiors planners conglomerate",
    database="hrp1_db"
)
cursor = db.cursor()
query = "SELECT product_name, SUM(price * quantity) AS total_revenue FROM Revenue GROUP BY product_name ORDER BY total_revenue DESC LIMIT 5"
cursor.execute(query)

products, revenues = [], []
for row in cursor:
    product, revenue = row
    products.append(product)
    revenues.append(revenue)

cursor.close()
db.close()

plt.figure(figsize=(12, 8))
plt.barh(products, revenues, color='orange')
plt.xlabel('Total Revenue')
plt.ylabel('Product Name')
plt.title('Top 5 Highest Revenue-Generating Products')
plt.gca().invert_yaxis()  
plt.show()
```
