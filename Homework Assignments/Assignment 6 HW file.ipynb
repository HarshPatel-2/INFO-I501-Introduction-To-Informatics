```python
try:
    key = input("Enter a key: ")
    dictionary_string = input("Enter a dictionary (in the format {'key': 'value'}): ")
    dictionary = eval(dictionary_string)

    if not isinstance(dictionary, dict):
        raise ValueError("Input is not a dictionary")
    value = dictionary.get(key)

    if value is not None:
        print(f"The value associated with the key '{key}' is: {value}")
    else:
        print(f"Key '{key}' not found in the dictionary")

except ValueError as ve:
    print(f"ValueError: {ve}")
except KeyError as ke:
    print(f"KeyError: '{key}' not found in the dictionary")
except NameError as ne:
    print(f"NameError: {ne}")
```

```python
class NumberList: 
    def __init__(self):
        self.elements =[]

    def add_element(self, number):
        self.elements.append(number)

    def sort_descending_list(self):
        self.elements.sort(reverse=True)

    def sort_ascending_list(self):
        self.elements.sort()

number_list = NumberList()
number_list.add_element(15)
number_list.add_element(21)
number_list.add_element(11)
number_list.add_element(14)
number_list.add_element(32)
number_list.add_element(67)

print("The Orginal list:", number_list.elements)
number_list.sort_descending_list()
print("The list sorted in decscending order:", number_list.elements)
number_list.sort_ascending_list()
print("The list sorted in ascending order:", number_list.elements)
```

```python
import pdb

class ElementWiseAddition:
    def __init__(self, list1, list2):
        if len(list1) != len(list2):
            raise ValueError("Both of the lists must be of the same length.")

        self.result = [a + b for a, b in zip(list1, list2)]

        pdb.set_trace()

list1 = [5, 10, 15, 20, 25]
list2 = [25, 20, 15, 10, 5]

try:
    ewa = ElementWiseAddition(list1, list2)
    print("The result of the Element Wise addition result:", ewa.result)
except ValueError as ve:
    print(ve)
```

```python
from abc import ABC, abstractmethod

class Cricket(ABC):
    @abstractmethod
    def details(self):
        pass

class Matches(Cricket):
    def details(self, match_type):
        return f"Match Type: {match_type}"

    def t20(self):
        return "This is a T-20 match."

    def test(self):
        return "This is a Test match."

class Teams(Cricket):
    def details(self, team_type):
        return f"Team Type: {team_type}"

    def national(self):
        return "This is a National team."

    def domestic(self):
        return "This is a Domestic team."

class Skills(Cricket):
    def details(self, skill_type):
        return f"Skill Type: {skill_type}"

    def batting(self):
        return "This player is good at batting."

    def fielding(self):
        return "This player is phenomenal at fielding."

    def bowling(self):
        return "This player is skilled at bowling."

class Equipments(Cricket):
    def details(self, equipment_type):
        return f"Equipment Type: {equipment_type}"

    def bat(self):
        return "This equipment is a cricket bat."

    def ball(self):
        return "This equipment is a cricket ball."

    def gloves(self):
        return "These equipment are cricket gloves."

match = Matches()
print(match.details("T-20"))
print(match.t20())
print(match.test())

team = Teams()
print(team.details("National"))
print(team.national())
print(team.domestic())

skill = Skills()
print(skill.details("Batting"))
print(skill.batting())
print(skill.fielding())
print(skill.bowling())

equipment = Equipments()
print(equipment.details("Bat"))
print(equipment.bat())
print(equipment.ball())
print(equipment.gloves())
```
