```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

num_points = 75
x_data = np.sort(np.random.uniform(-10, 10, num_points))
y_data_gaussian = np.random.normal(0, 1, num_points)

def polynomial_equation(x):
    return 4 * x**5 + 5 * x**3 - 3 * x**2 + 5 * x + 8

y_data_polynomial = polynomial_equation(x_data)
y_data_combined = y_data_gaussian + y_data_polynomial

degree = 5
coefficients = np.polyfit(x_data, y_data_combined, degree)
y_fitted = np.polyval(coefficients, x_data)

plt.scatter(x_data, y_data_combined, label='Generated Data')
plt.plot(x_data, y_fitted, color='red', label='Fitted Curve')
plt.title('Least-Squares Polynomial Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
import warnings

file_path = 'Election_data.csv'
df = pd.read_csv(file_path)

candidate_a_votes = df['Candidate A Votes']
total_votes = df['Total Votes']

best_r2_score = -1
best_model_degree = -1

warnings.filterwarnings("ignore", category=np.RankWarning)

for degree in range(1, 11):
    fit_coefficients = np.polyfit(candidate_a_votes, total_votes, degree)
    total_votes_predicted = np.polyval(fit_coefficients, candidate_a_votes)
    r2 = r2_score(total_votes, total_votes_predicted)
    
    if r2 > best_r2_score:
        best_r2_score = r2
        best_model_degree = degree

print(f"Best Model Degree: {best_model_degree}")
print(f"R-squared Value for Best Model: {best_r2_score}")
```

```python
import pandas as pd
from scipy.stats import ttest_ind

file_path = "Election_data.csv"
election_data = pd.read_csv(file_path)

candidate_a_votes = election_data["Candidate A Votes"]
candidate_b_votes = election_data["Candidate B Votes"]

t_statistic, p_value = ttest_ind(candidate_a_votes, candidate_b_votes)

alpha = 0.05

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("The means of Candidate A votes and Candidate B votes are significantly different.")
else:
    print("The means of Candidate A votes and Candidate B votes are not significantly different.")
```
